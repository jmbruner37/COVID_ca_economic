```python
import pandas as pd
```

```python
covid = pd.read_csv("../Raw_Data/CA_County_Covid_Cases.csv")
ue = pd.read_csv("../Raw_Data/CA_Unemployment.csv")
home = pd.read_csv("../Raw_Data/Zillow_House_Prices.csv")
```

```python
covid["month"] = pd.Series(covid["date"]).str.split("/").str[0]
covid.head()
```

```python
covid = covid[(covid.county != "Unassigned") & (covid.county != "Out Of Country")]
covid.head(5)
```

```python
covid_gb = covid.groupby(["county", "month"]).sum()
covid_gb = covid_gb[["newcountconfirmed", "newcountdeaths"]]
covid_gb = covid_gb.rename(columns={"newcountconfirmed":"new_cases", "newcountdeaths":"new_deaths"})
covid_gb.head()
```

```python
covid_gb.to_csv("../Raw_Data/Covid_Transformed.csv")
```

```python
ue.head()
```

```python
ue = ue.loc[(ue['Area Type '] == 'County') & (ue['Year '] == 2020)]
ue.head()
```

```python
ue["month"] = pd.Series(ue["Date"]).str.split("/").str[0]
ue.head()
```

```python
ue["Area Name "] = pd.Series(ue["Area Name "]).str.split(" ").str[:-1].apply(' '.join)
ue = ue.rename(columns={"Unemployment Rate ":"unemployment_rate", "Area Name ":"county", "Employment ":"employment", 
                        "Unemployment ":"unemployment", "Labor Force":"labor_force"})
ue.head()
```

```python
ue_gb = ue.groupby(["county", "month"]).sum()
ue_gb = ue_gb[["labor_force", "employment", "unemployment","unemployment_rate"]]
ue_gb.head()
```

```python
ue_gb.to_csv("../Raw_Data/Unempolyment_Transformed.csv")
```

```python
home = home.loc[(home['State'] == 'CA')]
home.head()
```

```python
home = home[["RegionName", "State", "2/29/2020", "3/31/2020", 
             "4/30/2020", "5/31/2020", "6/30/2020", "7/31/2020", "8/31/2020"]]
home["county"] = pd.Series(home["RegionName"]).str.split(" ").str[:-1].apply(' '.join)
home.head()
```

```python
home = home.melt(id_vars=["RegionName", "State", "county"], 
        var_name="date", 
        value_name="price")
home.head()
```

```python
home = home[["county", "date", "price"]]
home.head()
```

```python
home["month"] = pd.Series(home["date"]).str.split("/").str[0]
home = home[["county", "month", "price"]]
home.head(700)
```

```python
home.to_csv("../Raw_Data/Home_Prices_Transformed.csv")
```
