```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
covid_cases_raw_file = "../Raw_Data/CA_County_Covid_Cases.csv"
covid_cases_raw = pd.read_csv(covid_cases_raw_file)
covid_cases_raw.head()
```

```python
new_covid_cases_raw = covid_cases_raw[['county', 'totalcountconfirmed', 'date']].copy()
new_covid_cases_raw.head()
new_covid_cases_with_dates = new_covid_cases_raw.rename(columns = {"county": "County",
                                         "totalcountconfirmed": "Total Case Count",
                                         "date": "Date"})
new_covid_cases_with_dates.head()
#list(new_covid_cases_raw.columns)
```

```python
new_covid_cases_with_dates.to_csv("../Raw_Data/Covid_Cases_with_Dates.csv")
```

```python
unemployment_numbers_file = "../Raw_Data/CA_Unemployment.csv"
unemployment_numbers_raw = pd.read_csv(unemployment_numbers_file)
unemployment_numbers_raw.head(20)
```

```python
new_unemployment_numbers = unemployment_numbers_raw.loc[(unemployment_numbers_raw['Area Type '] == 'County') & (unemployment_numbers_raw['Year '] == 2020) & (unemployment_numbers_raw['Month'] != 'January') & (unemployment_numbers_raw['Month'] != 'February')]
new_unemployment_numbers.head()
list(new_unemployment_numbers.columns)
```

```python
# unemployment_columns = ["Area Name", "Date", "Unemployment"]
# unemployment_transformed = new_unemployment_numbers[unemployment_columns].copy()
# unemployment_transformed = unemployment_transformed.rename(columns = {"Area Name": "County",
#                                                                      "Date": "Date",
#                                                                      "Unemployment": "Unemployment"})
# unemployment_transformed.head()

unemployment_transformed= new_unemployment_numbers[['Area Name ', 'Date', 'Unemployment ']].copy()
unemployment_transformed = unemployment_transformed.rename(columns = {"Area Name": "County",
                                                                     "date": "Date",
                                                                     "Unemployment": "Unemployment"})
unemployment_transformed.head()
```

```python
unemployment_transformed.to_csv("../Raw_Data/Unemployment_with_Dates.csv")
```

```python
home_price_file = "../Raw_Data/Zillow_House_Prices.csv"
home_price_raw = pd.read_csv(home_price_file)
home_price_raw.head(20)
```

```python
new_home_price_raw = home_price_raw.loc[(home_price_raw['State'] == 'CA')]
new_home_price_raw.head()
new_home_price_raw_copy = new_home_price_raw.copy()
```

```python
home = new_home_price_raw_copy[["RegionName", "State", "2/29/2020", "3/31/2020", 
             "4/30/2020", "5/31/2020", "6/30/2020", "7/31/2020", "8/31/2020"]]
#home["county"] = pd.Series(home["RegionName"]).str.split(" ").str[:-1].apply(' '.join)
home.head()
```

```python
home = home.melt(id_vars=["RegionName", "State"], 
        var_name="date", 
        value_name="price")
home.head()
```

```python
home_transformed = home.rename(columns = {"RegionName": "County",
                                         "State": "State",
                                         "date": "Date",
                                         "price": "Price"})
home_transformed.head()
```

```python
home_transformed.to_csv("../Raw_Data/Home_Prices_with_Dates.csv")
```
